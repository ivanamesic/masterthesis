```python
import json
import requests
import pandas as pd
import time
```

```python
url = "https://api.tzkt.io/v1/operations/transactions/count"
params = {'limit': 10000,
          'level.ge': 251110, 
          'level.lt':251310}


start_time = time.time()

response = json.loads(requests.get(url = url, params = params).text)
print(json.dumps(response, indent=2))

print("--- %s seconds ---" % (time.time() - start_time))
```
