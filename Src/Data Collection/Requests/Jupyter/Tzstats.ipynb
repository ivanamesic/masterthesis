```python
import pandas as pd
import requests
from datetime import datetime
import json

final_dir = "../../../Data/Tezos/Final/"
```

```python
tz_start_date = "2018-07-03"
tz_end_date = "2023-06-27"

ts_start = int(datetime.timestamp(pd.to_datetime(tz_start_date))*1000)
ts_end = int(datetime.timestamp(pd.to_datetime(tz_end_date))*1000)


params = {
    "start_date": ts_start,
    "end_date": ts_end,
    "limit": 50000,
    "collapse": "1d"
}
```

## Block aggregated data

```python
url = "https://api.tzstats.com/series/block"


columns_names_blocks  = ["time", "count", "n_endorsed_slots", "n_ops_applied", "n_ops_failed", "n_calls", "n_rollup_calls", "n_events", "n_tx", "n_tickets", "volume", "fee", "reward", "deposit", \
     "activated_supply",  "minted_supply",   "burned_supply",   "n_new_accounts", "n_new_contracts", "n_cleared_accounts", "n_funded_accounts", "gas_used", "storage_paid"]

params["columns"] = ",".join(columns_names_blocks)
```

```python
response = requests.get(url=url, params=params)
result = json.loads(response.text)
```

```python
df_block = pd.DataFrame.from_dict(result)
df_block.columns = columns_names_blocks
```

```python
df_block["date"] = df_block.time.apply(lambda x: datetime.fromtimestamp(x/1000))
df_block
```

## Chain aggregated data

```python
url = "https://api.tzstats.com/series/chain"

columns_chain = ["time", "total_accounts", "total_contracts", "total_contract_calls", "total_delegations", "total_transactions", "total_proposals", "total_ballots", \
 "funded_accounts", "dust_accounts", "ghost_accounts", "unclaimed_accounts", "total_delegators", "active_delegators", "inactive_delegators", "total_bakers", "active_bakers", \
    "inactive_bakers"]


params["columns"] = ",".join(columns_chain)
```

```python
response = requests.get(url=url, params=params)
result = json.loads(response.text)
```

```python
df_chain = pd.DataFrame.from_dict(result)
df_chain.columns = columns_chain
```

```python
df_chain["date"] = df_chain.time.apply(lambda x: datetime.fromtimestamp(x/1000))
df_chain = df_chain[["date"] + columns_chain]
df_chain
```

## Supply data

```python

columns_supply = ["height", "cycle", "time", "total", "activated", "unclaimed", "circulating", "liquid", "delegated", "staking",  "active_delegated", "active_staking", \
                  "inactive_delegated", "inactive_staking", "minted", "burned", "frozen"]

params = {
    "start_date": ts_start,
    "end_date": ts_end,
    "limit": 50000,
    "collapse": "1d", 
    "columns": ",".join(columns_supply)
}

url = "https://api.tzstats.com/series/supply"


response = requests.get(url=url, params=params)
result = json.loads(response.text)

df_supply = pd.DataFrame.from_dict(result)
df_supply.columns = columns_supply

df_supply["date"] = df_supply.time.apply(lambda x: datetime.fromtimestamp(x/1000))
df_supply = df_supply[["date"]+columns_supply]
```

```python
df_supply.to_csv(final_dir + "Tzstats_supply_daily.csv", index=False)
```

## Divide data into groups

```python
# Accounts data
df_account_block = df_block[["date", "n_new_accounts", "n_cleared_accounts", "n_funded_accounts"]]
df_account_chain = df_chain[["date", "total_accounts", "funded_accounts", "dust_accounts", "ghost_accounts", "unclaimed_accounts"]]

df_account = pd.merge(df_account_block, df_account_chain, on="date", how="left")
df_account.columns = ["date", "new_accounts", "cleared_accounts", "funded_accounts", "total_accounts", "total_funded_accounts", "total_dust_accounts", "total_ghost_accounts", "total_unclaimed_accounts"]

df_account.to_csv(final_dir + "Tzstats_accounts_daily.csv", index= False)
```

```python
#Contracts data
df_contract_block = df_block[["date", "n_calls",  "n_new_contracts"]]
df_contracts_chain = df_chain[["date", "total_contracts", "total_contract_calls"]]
df_contracts = pd.merge(df_contract_block, df_contracts_chain, on="date", how="left")

df_contracts.to_csv(final_dir + "tzstats_contracts_daily.csv", index=False)
```

```python
# Transaction data

df_transactions_block = df_block[["date", "n_ops_applied", "n_ops_failed", "n_tx",  "volume"]]
df_transactions_chain = df_chain[["date", "total_transactions"]]

df_transaction = pd.merge(df_transactions_block, df_transactions_chain, on="date", how="left")
df_transaction.to_csv(final_dir + "Tzstats_transaction_daily.csv", index = False)
```

```python
# Baking data

df_baking_chain = df_chain[["date", "total_bakers", "active_bakers", "inactive_bakers", "total_delegators", "active_delegators", "inactive_delegators"]]
df_baking_supply = df_supply[["date",  "delegated", "staking",  "active_delegated", "active_staking",  "inactive_delegated", "inactive_staking"]]

df_baking = pd.merge(df_baking_chain, df_baking_supply, on="date", how="left")

df_baking.to_csv(final_dir + "Tzstats_baking_daily.csv", index=False)
```

```python
# Voting data
voting = df_chain[["date", "total_proposals", "total_ballots"]]
voting.to_csv(final_dir + "Tzstats_voting_daily.csv", index = False)
```
