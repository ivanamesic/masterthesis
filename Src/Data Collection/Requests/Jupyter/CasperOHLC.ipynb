```python
import requests, json
import pandas as pd
from datetime import datetime
#data taken with free api plan on twelvedata

root_url = "https://api.twelvedata.com/"
api_key = "6c15cc21bea546dbab54c8507d288870"
```

```python
url = root_url
params = {
    "symbol": "CSPR/USD",
    "interval":"1day",
    "apikey": api_key,
    "outputsize":"5000"
}

params_string = ""
for k, v in params.items():
    params_string+= k+"="+v+'&'
params_string = params_string[:-1]

url = root_url + "time_series?" + params_string
```

```python
resp = requests.get(url = url)
result = json.loads(resp.text)
```

```python
df_ohlc = pd.DataFrame.from_dict(result["values"])
df_ohlc
```
